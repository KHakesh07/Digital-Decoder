```python
#REQUIRED LIBRARIRES
from keras.datasets import mnist
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#DIVIDING THE DATASET
(data_images, data_labels),(test_images,test_labels) = mnist.load_data()
```

```python
#VISUALIZING THE FEW IMAGES OF DATASET
import matplotlib.pyplot as plt
for i in range(20,25):
  print(plt.matshow(data_images[i]))
```

```python
#CALLING THE FEW LABELS THAT MATCH THE ABOVE IMAGES
data_labels[20:25]
```

```python
#NORMALIZING THE DATASET
data_images = data_images.astype('float32')/255
```

```python
#SPLITING THE DATASET INTO TRAINING, VALIDATION DATA
X_train, X_val, y_train, y_val = train_test_split(data_images, data_labels, test_size=0.1)
```

```python
#RESHAPING THE DATA
X_train = X_train.reshape(-1,28*28)
X_val = X_val.reshape(-1,28*28)
test_images = test_images.reshape(-1,28*28)
```

```python
#IMPORTING THE LOGISTIC REGRESSION MODEL
from sklearn.linear_model import LogisticRegression
```

```python
#THE LOGISTIC REGRESSION MODEL
model = LogisticRegression()
```

```python
#TRAINING THE MODEL
model.fit(X_train,y_train)
```

```python
#COEFFICIENTS
model.coef_[0]
```

```python
#INTERCEPT
model.intercept_[0]
```

```python
#MODEL PREDICITON BASED ON VALIDATION DATA
val_predict = model.predict(X_val)
```

```python
#CONFIRMING THAT PREDICTED DATA IS EQUAL TO LABEL DATA
val_predict == y_val
```

```python
#FINDING THE ACCURACY FOR VALLIDATION DATA
val_accuracy = accuracy_score(y_val, val_predict)
```

```python
#ACCURACY
val_accuracy
```

```python
#PREDICTING THE TEST DATA
test_predict = model.predict(test_images)
```

```python
#CHECKING IF PREDICTED DATA == LABELED DATA
test_predict == test_labels
```

```python
#FINDING ACCURACY FOR TEST DATA
test_accuracy = accuracy_score(test_labels, test_predict)
```

```python
test_accuracy
```
